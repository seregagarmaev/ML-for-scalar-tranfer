# Description

In this notebook we prepare data for the TBNNs algorithm.

# Import

```python
import pandas as pd
from scipy.interpolate import interp1d
```

# Re_tau = 180

```python
data = pd.read_csv('data/DNS/Re_tau_180/data.csv')
```

```python
data = data.rename(columns={'ut+': 'ut', 'vt+': 'vt'})
```

```python
nu = 0.00035
u_tau = 0.063012

data['y'] = data['y+'] * nu / u_tau
```

```python
data.to_csv('data/DNS/Re_tau_180/processed_data.csv', index=False)
```

```python
data
```

# Re_tau = 395

```python
data1 = pd.read_csv('data/DNS/Re_tau_395/data1.csv')
data2 = pd.read_csv('data/DNS/Re_tau_395/data2.csv')
```

```python
f = interp1d(data2['y+'], data2['-vt+'], kind='cubic', fill_value='extrapolate')
```

```python
data1['vt'] = f(data1['y+'])
```

```python
data1 = data1.rename(columns={'ut+': 'ut'})
```

```python
nu = 0.000145
u_tau = 0.0574282

data1['y'] = data1['y+'] * nu / u_tau
```

```python
data1.to_csv('data/DNS/Re_tau_395/processed_data.csv', index=False)
```

```python
data1
```
